```python
import pandas as pd

# Load the source CSV (the one you want to copy the column from)
source_csv1= '/content/drive/MyDrive/Colab Notebooks/targets/581885_40_2023-03-08-135646-0000.csv'
source_df1 = pd.read_csv(source_csv1)

# Load the target CSV (the one you want to add the column to)
target_csv1 = '/content/drive/MyDrive/Colab Notebooks/copy-columns/581885_40_2023-03-08-135646-0000.csv'
target_df1 = pd.read_csv(target_csv1)

# Specify the column you want to copy
column_to_copy = 'grooming'

# Copy the column from source_df to target_df
target_df1[column_to_copy] = source_df1[column_to_copy]

# Save the modified target DataFrame to a new CSV file
output_csv1 = '/content/drive/MyDrive/Colab Notebooks/output/581885_40_2023-03-08-135646-0000.csv'
target_df1.to_csv(output_csv1, index=False)

```
