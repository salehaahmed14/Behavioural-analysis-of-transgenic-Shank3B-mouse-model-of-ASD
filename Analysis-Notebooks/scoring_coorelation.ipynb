```python
#necessary imports
import numpy as np
import pandas as pd
from sklearn.metrics import cohen_kappa_score,matthews_corrcoef,adjusted_rand_score
```

```python
#import data
scorer1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/scored_by_saleha.csv')['grooming'].values
scorer2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/scored_by_zijia.csv')['grooming'].values
```

```python
kappa = cohen_kappa_score(scorer1, scorer2)
#print the coefficient
print(f'cohens kappa: {kappa}')
```

```python
mcc = matthews_corrcoef(scorer1, scorer2)
#print the coefficient
print(f'matthews coorelation coefficient: {mcc}')
```

Cohen suggested the Kappa result be interpreted as follows: values ≤ 0 as indicating no agreement and 0.01–0.20 as none to slight, 0.21–0.40 as fair, 0.41– 0.60 as moderate, 0.61–0.80 as substantial, and 0.81–1.00 as almost perfect agreement. However, this interpretation allows for very little agreement among raters to be described as “substantial”. For percent agreement, 61% agreement can immediately be seen as problematic. Almost 40% of the data in the dataset represent faulty data.

**Model Scoring and Human Scoring (frame by frame comparison)**

```python
#import data
scored_by_zijia = pd.read_csv('/content/drive/MyDrive/comparison--scoring/581885_f42_wt_2023-03-08-151135-0000.csv')['grooming'].values
scored_by_model = pd.read_csv('/content/drive/MyDrive/comparison--scoring/581885_f42_wt_2023-03-08-151135-0000-m.csv')['grooming'].values
```

```python
ari = adjusted_rand_score(scored_by_zijia, scored_by_model)
print(f'The Adjusted Rand Index (ARI) is: {ari:.4f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# Assuming 'scored_by_zijia' and 'scored_by_model' are already defined
# Confusion matrix
cm = confusion_matrix(scored_by_zijia, scored_by_model)

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', linewidths=1.5, linecolor='black')

# Customizing labels and title with bold font
plt.xlabel('Predicted Labels', fontsize=14, fontweight='bold')
plt.ylabel('True Labels', fontsize=14, fontweight='bold')
plt.title('Confusion Matrix of Grooming Behavior Classification', fontsize=16, fontweight='bold')

# Customizing tick labels with bold font
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.savefig('/content/drive/MyDrive/results/confusion_report.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Calculate classification report
report = classification_report(scored_by_zijia, scored_by_model, target_names=['Not Grooming', 'Grooming'])

# Print the classification report
print(report)
```

```python
kappa = cohen_kappa_score(scored_by_zijia, scored_by_model)
#print the coefficient
print(f'cohens kappa: {kappa}')
```
