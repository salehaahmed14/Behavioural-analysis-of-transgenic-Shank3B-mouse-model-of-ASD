```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_grooming_bout_durations(file_path, frame_rate=30.09):
    """
    Function to plot a histogram of grooming bout durations.

    Parameters:
    file_path: path to the csv file containing grooming data.
    frame_rate: frame rate of the video (default is 30.09).
    """
    #load the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    #display the first few rows of the DataFrame to understand its structure
    print(df.head())

    #1 indicates grooming and 0 indicates no grooming

    #identify grooming bouts
    df['bout_id'] = (df['grooming'] != df['grooming'].shift()).cumsum()

    #filter out non-grooming bouts
    grooming_bouts = df[df['grooming'] == 1]

    #calculate the duration of each grooming bout in frames
    bout_durations_frames = grooming_bouts.groupby('bout_id').size()

    #convert frame durations to seconds using the frame rate
    bout_durations_seconds = bout_durations_frames / frame_rate

    #plot the histogram
    plt.figure(figsize=(10, 6))
    plt.hist(bout_durations_seconds, bins=25, edgecolor='black')
    plt.xlabel('Bout Duration (seconds)')
    plt.ylabel('Frequency')
    plt.title('Histogram of Grooming Bout Durations')
    plt.show()
```

```python
plot_grooming_bout_durations('/content/drive/MyDrive/comparison--scoring/I581886_m25_wt_2023-03-08-121136-0000.csv')
```
