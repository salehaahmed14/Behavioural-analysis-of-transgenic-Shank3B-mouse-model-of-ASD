```python
import pandas as pd

# Load the CSV file
file_path = '/content/drive/MyDrive/Colab Notebooks/data_summary_20240808170041.csv'
data = pd.read_csv(file_path)

# Display the contents of the file
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract group information from the 'Video' column
data['Group'] = data['Video'].apply(lambda x: 'homo' if 'homo' in x else ('wt' if 'wt' in x else 'het'))

# Calculate the average grooming times for each group
group_means = data.groupby('Group')['Value'].mean()

# Create a boxplot for the grooming times by group
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Value', data=data, palette=["blue", "orange", "red"])
plt.title('Grooming Times by Group')
plt.xlabel('Group')
plt.ylabel('Total Grooming Time (s)')
plt.show()

# Output the calculated means
group_means
```
